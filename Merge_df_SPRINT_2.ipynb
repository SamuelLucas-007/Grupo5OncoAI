#1.5 Pré-Processamento

## Utilizando Simple Imputer para tratar missing numbers.

```python
from sklearn.impute import SimpleImputer
```

```python
imp_mean = SimpleImputer(missing_values=np.nan, strategy='mean')
imp_median = SimpleImputer(missing_values=np.nan, strategy='median')
imp_most_frequent = SimpleImputer(missing_values=np.nan, strategy='most_frequent')
imp_constant = SimpleImputer(missing_values=np.nan, strategy='constant', fill_value="Nao informado")
imp_constant2 = SimpleImputer(missing_values=np.nan, strategy='constant', fill_value="Nao realizado")
```

### Colunas que serão utilizadas a mediana como estratégia.

```python
df_original['idade_no_primeiro_diagnostico'] = imp_median.fit_transform(df_original[['idade_no_primeiro_diagnostico']])
df_original['subtipo_tumoral'] = imp_median.fit_transform(df_original[['subtipo_tumoral']])
df_original['indice_h_receptor_de_progesterona'] = imp_median.fit_transform(df_original[['indice_h_receptor_de_progesterona']])
df_original['ki67_percentage'] = imp_median.fit_transform(df_original[['ki67_percentage']])
df_original['grau_histologico'] = imp_median.fit_transform(df_original[['grau_histologico']])
```

### Colunas que serão utilizadas a most_frequent como estratégia.

```python
df_original['sexo'] = imp_most_frequent.fit_transform(df_original[['sexo']])
```

### Colunas que serão utilizadas constantes como estratégia.

```python
df_original['ja_ficou_gravida'] = imp_constant.fit_transform(df_original['ja_ficou_gravida'].values.reshape(-1,1))
df_original['realiza_atividades_fisicas'] = imp_constant.fit_transform(df_original['realiza_atividades_fisicas'].values.reshape(-1,1))
df_original['ja_usou_drogas'] = imp_constant.fit_transform(df_original['ja_usou_drogas'].values.reshape(-1,1))
df_original['consumo_de_alcool'] = imp_constant.fit_transform(df_original['consumo_de_alcool'].values.reshape(-1,1))
df_original['consumo_de_tabaco'] = imp_constant.fit_transform(df_original['consumo_de_tabaco'].values.reshape(-1,1))
df_original['possui_historico_familiar_de_cancer'] = imp_constant.fit_transform(df_original['possui_historico_familiar_de_cancer'].values.reshape(-1,1))
df_original['classificacao_tnm_patologico_n'] = imp_constant.fit_transform(df_original['classificacao_tnm_patologico_n'].values.reshape(-1,1))
df_original['classificacao_tnm_patologico_t'] = imp_constant.fit_transform(df_original['classificacao_tnm_patologico_t'].values.reshape(-1,1))
df_original['classificacao_tnm_patologico_t'] = imp_constant.fit_transform(df_original['classificacao_tnm_patologico_t'].values.reshape(-1,1))
```

```python
df_original['receptor_de_estrogenio'] = imp_constant2.fit_transform(df_original['receptor_de_estrogenio'].values.reshape(-1,1))
df_original['receptor_de_progesterona'] = imp_constant2.fit_transform(df_original['receptor_de_progesterona'].values.reshape(-1,1))
df_original['ki67_maior_14_percentage'] = imp_constant2.fit_transform(df_original['ki67_maior_14_percentage'].values.reshape(-1,1))
df_original['receptor_de_progesterona_quantificacao_percentage'] = imp_constant2.fit_transform(df_original['receptor_de_progesterona_quantificacao_percentage'].values.reshape(-1,1))
df_original['receptor_de_estrogenio_quantificacao_percentage'] = imp_constant2.fit_transform(df_original['receptor_de_estrogenio_quantificacao_percentage'].values.reshape(-1,1))
df_original['her2_por_ihc'] = imp_constant2.fit_transform(df_original['her2_por_ihc'].values.reshape(-1,1))
df_original['her2_por_fish'] = imp_constant2.fit_transform(df_original['her2_por_fish'].values.reshape(-1,1))
```

## Codificação correta das variáveis categóricas e normalização das variáveis numéricas.
